```python
# read the image file
img = cv2.imread('input.png')
```

```python
print(type(img))
print(img.ndim)
rowNum, columnNum, pixNum = img.shape
print(img.shape)
print(rowNum)
print(columnNum)
print(img.size)
```

```python
#operation 1
output1Image = img.copy()

red = np.array((255, 0, 0))
green = np.array((0, 255, 0))
blue = np.array((0, 0, 255))

# note: when the points with which the pixel has least distance include red, we make the pixel red (highest priority);
# when the points with which the pixel has least distance include green but not red, we make the pixel green (second priority);
for row in output1Image:
  for pixel in row:
    # using linalg.norm() to calculate Euclidean distance
    dist1 = np.linalg.norm(red - pixel)
    leastDistance = dist1
    targetedColor = red

    dist2 = np.linalg.norm(green - pixel)
    if dist2 < leastDistance:
        leastDistance = dist2
        targetedColor = green

    dist3 = np.linalg.norm(blue - pixel)
    if dist3 < leastDistance:
        targetedColor = blue
    pixel[0] = targetedColor[0]
    pixel[1] = targetedColor[1]
    pixel[2] = targetedColor[2]

cv2_imshow(output1Image)
cv2.imwrite('output1.png', output1Image)
```

```python
#operation 2
output2Image = img.copy()

height, width, pixelSize = output2Image.shape

dark = np.array((0, 0, 0))

output2Image[height//2 - 25 : height//2 + 25, width//2 - 25 : width//2 + 25] = dark

cv2_imshow(output2Image)
cv2.imwrite('output2.png', output2Image)
```
