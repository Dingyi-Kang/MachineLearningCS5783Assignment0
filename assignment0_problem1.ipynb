```python
#open text file in read mode
f = open("test.txt", "r")

#read whole file to a string
content = f.read()

f.close()
```

```python
letters = 'abcdefghijklmnopqrstuvwxyz'

## initiate the letter frequency dictionary
letter_frequency = dict([(l, 0) for l in letters])

content = content.lower()

##iterate each characters in the string of the file
for i in content:
  if i.isalpha():
    if i in letter_frequency:
      letter_frequency[i] += 1
    else:
      letter_frequency[i] = 1
## alternative of this above if-else statement, we get use dict.get(i, defaultValue) function
print("The histogram as a dictionary is: " + str(letter_frequency))
print("The number of keys in this dictionary is " + str(len(letter_frequency.keys())))
```

```python
##initiate an numpy array of size 26 with default value of 0
letterFreq_array = np.zeros(26, dtype=int)

for key, value in letter_frequency.items():
  #store the letter frequency based the ASCII value of the letter. For example, the ASCII value of 'a' is 97, hence we store it at the index of (97-97), namely index of 0 in the array
  letterFreq_array[ord(key)-97] = value

print("The histogram as an array is: " + str(letterFreq_array))
print("The length of this array is " + str(letterFreq_array.size))
```

```python
# The total number of letters appearing in the text
sum_letter = np.sum(letterFreq_array)
##print(sum_letter)

# create a gifure with certain size
fig = plt.figure(figsize = (18, 5))
# The bar graph represents normalized letters frequencies, with letters along the horizontal axis and the calculated
# letter frequencies as percentages as the bar height
x = list(letter_frequency.keys())
bars = plt.bar(x, [letter_frequency[l]/sum_letter * 100 for l in letters], width=0.8,
       color='g', alpha=0.5, align='center')
plt.xlabel("English letters")
plt.ylabel('Letter frequency (percentage/%)')
plt.title("Bar Graph of Normalized Letters Frequencies")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + 0.025, "{:.2f}%".format(yval))
plt.show()
```

```python
# create a gifure with certain size
fig = plt.figure(figsize = (18, 5))
# The bar graph represents unnormalized letters frequencies
bars = plt.bar(x, letter_frequency.values(), width=0.8,
       color='g', alpha=0.5, align='center')
plt.xlabel("English letters")
plt.ylabel('Letter frequency (times)')
plt.title("Bar Graph of Unnormalized Letters Frequencies")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + 0.025, yval)
plt.show()
```
